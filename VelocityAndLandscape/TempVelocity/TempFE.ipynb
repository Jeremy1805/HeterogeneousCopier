```python
df = pd.read_csv('TempTherm.FreeEnergyData.csv')
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 26}

matplotlib.rc('font', **font)
matplotlib.rcParams['figure.figsize'] = 6.4,4.8

ax = df.plot('l',linewidth=0.4)
plt.ylabel(r'$G_{\mathbf{x},l}-\langle G \rangle$')
plt.xlabel('$l$')
plt.legend(title='$\Delta G_{T,22}$')

plt.yticks(np.arange(-200, 201,100))
h, l = ax.get_legend_handles_labels()
plt.gca().set_ylim([-250, 220])
plt.gca().yaxis.set_label_coords(-0.15,0.5)

plt.legend().remove()
plt.tight_layout(pad = 0)
plt.gcf().set_size_inches(9.6, 5.4)

plt.savefig('FETempTherm.svg')
```
