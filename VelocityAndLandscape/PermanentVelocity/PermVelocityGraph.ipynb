```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
```

```python
session = WolframLanguageSession()
data = session.evaluate(wlexpr('Get["PermTherm.NormalizedVelocityData"]'))
```

```python
df = pd.DataFrame(data[0])
for i in range(1,len(data)):
    df = df.merge(pd.DataFrame(data[i]),on=0,suffixes=(None,'_'+str('i')))

df.columns = ['Gpol',2,4,6,8,10]
```

```python
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 26}

matplotlib.rc('font', **font)

df.plot('Gpol',xlabel='$\Delta G_{pol} - \Delta G_{pol,eq}$',ylabel='$ \dfrac{v}{v_{7.0}}$')
plt.ylabel('$ \overline{v}$',rotation=0)
plt.gca().yaxis.set_label_coords(-0.15,0.45)
plt.legend(title='$\Delta G_{T,22}$')
plt.xticks(np.arange(0, 7, 1.0))

plt.legend().remove()
plt.tight_layout(pad=0)
plt.gcf().set_size_inches(6.4, 5.4)
plt.tight_layout(pad=0)

plt.savefig('VelocityPermTherm.eps')
```
