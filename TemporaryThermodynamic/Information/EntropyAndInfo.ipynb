```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab

from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
```

```python
session = WolframLanguageSession()
data = session.evaluate(wlexpr('Get["Entropy/EntropyData"]+0.0'))
session.terminate()

df = pd.DataFrame(data[0])
for i in range(1,len(data)):
    df = df.merge(pd.DataFrame(data[i]),on=0,suffixes=(None,'_'+str('i')))

df.columns = ['pt',0,1,2,3,4,5,6]

font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 26}

matplotlib.rc('font', **font)

ax = df.plot('pt', marker='o')
plt.ylabel('$\max_{p_t}\eta_{Therm}$')
plt.xlabel('$\Delta G_{pol}$')
plt.xticks(np.arange(-0.6, 0.01, 0.2))
plt.gca().set_ylim([0, 0.7])

plt.legend().remove()
plt.tight_layout(pad = 0)
plt.gcf().set_size_inches(5.4, 5.4)
plt.tight_layout(pad = 0)

plt.savefig('ThermoIncorrect.ThermEfficiency.eps')
```
