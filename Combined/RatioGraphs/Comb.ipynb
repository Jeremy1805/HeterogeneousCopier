```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab

from wolframclient.evaluation import WolframLanguageSession
from wolframclient.language import wl, wlexpr
```

```python
session = WolframLanguageSession()
data = session.evaluate(wlexpr('Get["CombCorrect.Error.Log.Ratio"]'))
np.save("CombCorrect.Error.Log.Ratio",data)
session.terminate()
data = session.evaluate(wlexpr('Get["CombIncorrect.Error.Log.Ratio"]'))
np.save("CombIncorrect.Error.Log.Ratio",data)
session.terminate()
data = session.evaluate(wlexpr('Get["CombCorrect.Times.Log.Ratio"]'))
np.save("CombCorrect.Times.Log.Ratio",data)
session.terminate()
data = session.evaluate(wlexpr('Get["CombIncorrect.Times.Log.Ratio"]'))
np.save("CombIncorrect.Times.Log.Ratio",data)
session.terminate()
data = session.evaluate(wlexpr('Get["CombCorrect.Visits.Log.Ratio"]'))
np.save("CombCorrect.Visits.Log.Ratio",data)
session.terminate()
data = session.evaluate(wlexpr('Get["CombIncorrect.Visits.Log.Ratio"]'))
np.save("CombIncorrect.Visits.Log.Ratio",data)
session.terminate()
```

```python
data = np.load('CombCorrect.Error.Log.Ratio.npy')

df = pd.DataFrame(data[0])
for i in range(1,len(data)):
    df = df.merge(pd.DataFrame(data[i]),on=0,suffixes=(None,'_'+str('i')))

df.columns = ['pt',2,4,6,8,10]

font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 26}

matplotlib.rc('font', **font)

ax = df.plot('pt', marker='o')
plt.ylabel(r'$\ln{\dfrac{\langle \epsilon \rangle }{\langle \epsilon_I \rangle }}$')
plt.xlabel('$p_t$')
plt.legend(title='$\Delta G_{T,11}$')
plt.xticks(np.arange(0, 1.1, 0.2))

plt.legend().remove()
plt.tight_layout(pad = 0)
plt.gcf().set_size_inches(5.4, 5.4)
plt.tight_layout(pad = 0)

plt.savefig('CombCorrect.Error.eps')
```
